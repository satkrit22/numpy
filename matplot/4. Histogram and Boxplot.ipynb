# **Making Histogram**

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data=[1,3,3,3,3,9,9,4,4,8,8,8,6,7]   #creating a data
```

```python
plt.hist(data)   # Plotting a histogram with hist function
plt.show()
```

## **Adding with the bins**

```python
plt.hist(data, color="b", bins=3)   # Plotting a histogram with hist function and that bins only make the three box bins if bins=5 there there will be 5 boxes
plt.show()
```

# **Working with a real world data**

```python
plt.subplot(1,4,1)
plt.hist(iris_df['sepal_length'],bins=30,color="g")

plt.subplot(1,4,2)
plt.hist(iris_df['petal_length'],bins=30,color="r")

plt.subplot(1,4,3)
plt.hist(iris_df['sepal_width'],bins=30,color="b")

plt.subplot(1,4,4)
plt.hist(iris_df['petal_width'],bins=30,color="y")
plt.show()
```
