# **Violin Plot**

```python
import matplotlib.pyplot as plt
```

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]
data=list([one,two,three])
plt.violinplot(data, showmedians=True)
plt.show()
```

## **Changing asthetics**

```python
import matplotlib.pyplot as plt

Fruits = ["apple", "Mangos", "Banana", "Litchi", "Orange"]
quantity = [53, 43, 12, 97, 80]

plt.pie(quantity, labels=Fruits, autopct="%0.1f%%", colors=["green", "yellow", "red", "orange", "pink"])
plt.show()
```

# **Donut Chart**

```python
Fruits = ["apple", "Mangos", "Banana", "Litchi", "Orange"]
quantity = [53, 43, 12, 97, 80]
```

```python
plt.pie(quantity, labels=Fruits, autopct="%0.1f%%", colors=["green", "yellow", "red", "orange", "pink"], radius=2)
plt.pie([5],colors="w",radius=1)
plt.show()
```
