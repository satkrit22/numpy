```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
df['Gender'].value_counts().plot(kind='bar',color=['blue','pink'])
plt.title("Distribution of the gender")
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
a=df['Gender'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(a, labels=a.index,autopct='%1.1f%%',colors=['blue','pink'],startangle=140)
plt.title("Distribution of the gender")
plt.show()
```
